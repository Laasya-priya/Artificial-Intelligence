# Room Cleaning Robot (Dynamic Control Task)

Consider a Room Cleaner Robot which cleans a room containing dirt. A room can be treated as a $m \times n$ grid with walls on all sides. 

```python
import random
m = int(input('no.of rows:'))                      #input no.of rows
n = int(input('no.of columns:'))                   #input no.of columns
d = [[0 for i in range(n)] for j in range(m)]       #creating environment
count=1
while(count<=10):
    x = random.randint(0,m-1)         
    y = random.randint(0,n-1)
    if (d[x][y]!=1):                  #adding unit dirt if there is no existing dirt
        d[x][y]=1
        count+=1
print(d)
def add_dirt(t):                      #adding t units of dirt
    t1=0
    while (t1<=t):
        x1 = random.randint(0,m-1)
        y1 = random.randint(0,n-1)
        d[x1][y1]+=1
        t1+=1
add_dirt(1000)                         #adding 1000 units of dirt
print(d)
```

(b) This is an agent which at each time instant $t$ observes its position $o_t =(x_t, y_t)$  (but not the dirt information in this location) and performs a random action $a_t \in \{up,down,right,left, pickDirt\}$. On hitting a wall or picking dirt from the current location $(x_t, y_t)$, the agent stays in the same position $(x_t, y_t)$, otherwise, its position changes according to the action. Every action is associated with a reward (or penalty) defined as follows.

   
$$r_t = R((x_t, y_t), a_t) = 
     \begin{cases}
       -1, &\text{if the agent tries to pick dirt and } (x_t, y_t) \text{ is a clean cell}\\
      -10, &\text{if the agent hits a wall}
       \\ 
       d, &\text{if } (x_t, y_t) \text{ has } d \text{ units of dirt and the agent picks it}\\
       0, &\text{otherwise }
     \end{cases}
$$



```python
a=["up","down","right","left","pick"]               #all the possible choices
def point(x,y):                                    #giving input as current location
    pt=0
    ac=random.choice(a)
    print(ac)
    if (x==0 and ac=="up") or (x==m-1 and ac=="down") or (y==0 and ac=="left")or(y==n-1 and ac=="right"):   #cases for hitting the wall
        pt=-10
    elif (ac=="up"):
        x=x-1
    elif(ac=="down"):
        x+=1
    elif(ac=="right"):
        y+=1
    elif(ac=="left"):
        y=y-1
    if ( ac == "pick"):               #picking dirt case
        if(d[x][y] > 0) : 
            pt = d[x][y]
            d[x][y]=0             #making dirt zero once dirt is picked
        else : pt = -1                
    return(x,y,pt)
d
```
