# Tic-Tac-Toe using Alpha-Beta Pruning

The Tic-Tac-Toe game starts on a 3x3 grid with two players "X" and "O" who take turns and play. The rules are as follows: each player gets a turn with player "X" (resp. "O") writing an "X" (resp. "O") in an empty cell of the grid. The game starts with the move of the "O" player. The first player to write on three horizontal or vertical or diagonal cells wins.

```python
max_depth2=0
leaves2=0
def alphabeta(board, depth, is_maximizing_player,alpha,beta):
    global max_depth2
    global leaves2
    depth=depth+1
    # Base case: evaluate the board if game is over or maximum depth is reached
    score = evaluate_board(board)
    if score is not None:
        leaves2=leaves2+1
        return score
    if is_maximizing_player:
        best_score = -float("inf")
        for row in range(3):
            for col in range(3):  #maximising
                if (board[row][col] ==" "):
                    board[row][col] = 'X'
                    score = alphabeta(board, depth, False,alpha,beta)
                    board[row][col] = " "
                    best_score = max(score, best_score)
                    alpha=max(alpha,best_score)
                    if beta<=alpha:
                        break
                    max_depth2=max(max_depth2,depth)
        return best_score
    else:
        best_score = float("inf")
        for row in range(3):
            for col in range(3):      #minimising
                if board[row][col] ==" ":
                    board[row][col] = 'O'
                    score = alphabeta(board, depth , True,alpha,beta)
                    board[row][col] = " "
                    best_score = min(score, best_score)
                    beta=min(beta,best_score)
                    if beta<=alpha:
                        break
                    max_depth2=max(max_depth2,depth)
        return best_score
def findBestMove_alphabeta(board):
    global max_depth2
    max_depth2=0
    bestScore = -float("inf")
    alpha=-float("inf")     #alpha as minus infinity
    beta=float("inf")      #beta as plus infinity
    bestMove=None
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'
                score = alphabeta(board, 1, False,alpha,beta)
                board[i][j] = ' '
                if score > bestScore:
                    bestScore = score
                    bestMove = (i, j)    #computing best move
    return bestMove
def playGame3():
    board=[[" "," "," "],[" "," "," "],[" "," "," "]]
    while True:
        # O's turn
        while True:
            num=int(input("enter the position of O:"))
            maps={1:(0,0),2:(0,1),3:(0,2),4:(1,0),5:(1,1),6:(1,2),7:(2,0),8:(2,1),9:(2,2)}
            (row,col) =maps[num] 
            if(board[row][col]=='X'):
                print("X is already present here")
            else:
                board[row][col] = 'O'
                print_board(board)
                break
        # Check if O won
        result = evaluate_board(board)
        if result == -1:
            print("Player O won")
            break

        # Check if game is a draw
        if (is_full(board)==1):
            print("Game is a draw")
            break

        # X's turn
        if(findBestMove_alphabeta != None):
            (row, col) = findBestMove_alphabeta(board)
            board[row][col] = 'X'
            print("Computer played at row = {}, col = {}".format(row, col))
            print_board(board)
            print("maximum depth in this game:",max_depth2)

        # Check if X won
        result = evaluate_board(board)
        if result == 1:
            print("Player X won")
            break

        # Check if game is a draw
        if (is_full(board)==1):
            print("Game is a draw")
            break
playGame3()
print("leaves explored and computed:",leaves2)
```
